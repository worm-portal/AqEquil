create_data0():

- the data0.csv file has two optional columns: 'dissrxn' and 'formula_ox'
- if data0.csv is missing the 'dissrxn' column, basis species are automatically selected to derive dissociation reactions.
- if data0.csv is missing the 'formula_ox' column, redox suppression is disallowed. Results in error. Raise exception if possible.
